# Fetch empirics

*use `eds-api` to det PV, WL and market prices. Productions are capacity--deflated*

```python
start, end = 2025, 2026

# solar
PV = VE(
    value_columns = ["SolarPowerLt10kW_MWh", "SolarPowerGe10Lt40kW_MWh", "SolarPowerGe40kW_MWh"],
    cap_column    = ["SolarPowerCapacity"],
    col_name      = "PV_VE",
    start         = start,
    end           = end,
    cache         = True,
    save_txt      = True,
)
```

```python
# onshore wind
WL = VE(
    value_columns = ["OnshoreWindLt50kW_MWh", "OnshoreWindGe50kW_MWh"],
    cap_column    = ["OnshoreWindCapacity"],
    col_name      = "WL_VE",
    start         = start,
    end           = end,
    cache         = True,
    save_txt      = True,
)
```

```python
# quantity and prices
wp_h, q_h, _ = get_wp_h(
    start         = start,
    end           = end,
    cache         = True,
    save_txt      = True,
)
```

*correct save structure* 

```python
pat_dir = pathlib.Path("variation_patterns")

for f in sorted(pat_dir.glob("*.txt")):
    
    lines = f.read_text().splitlines()
    assert len(lines) in (8760, 8784), f"{f.name}: unexpected length {len(lines)}"
    
    if len(lines) == 8784:
        f.write_text("\n".join(lines[:8760]) + "\n")
        print(f"{f.name}: trimmed 8784 -> 8760")
    else:
        print(f"{f.name}: already 8760, skipped")
```
